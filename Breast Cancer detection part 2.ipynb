```python
import pandas as pd
df=pd.read_csv("data.csv")
df.head()
```

```python
df.drop(df.columns[-1],axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_Y=LabelEncoder()
df.iloc[:,1]=labelencoder_Y.fit_transform(df.iloc[:,1].values)
df.head()
```

```python
Y=df.iloc[:,1]
X=df.iloc[:,2:]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.25,random_state=3)
X_train.head()
```

```python
Y_train
```

```python
X_train=pd.DataFrame(X_train)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
corr=X_train.corr(method='pearson')
sns.heatmap(corr,annot=True)
plt.show()
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train, 0.99)
len(set(corr_features))
print(corr_features)
```

```python
X_train.shape
```

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train.values)
X_train=scaler.transform(X_train.values)
X_test=scaler.transform(X_test.values)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
Yp_train=lr.predict(X_train)
from sklearn.metrics import r2_score
r2_score(Y_train,Yp_train)
```

```python
lr.fit(X_test,Y_test)
Yp_test=lr.predict(X_test)
r2_score(Y_test,Yp_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
log_r=LogisticRegression(random_state=0)
log_r.fit(X_train,Y_train)
Yp_log_r =log_r.predict(X_train)
print(accuracy_score(Y_train,Yp_log_r))
```
