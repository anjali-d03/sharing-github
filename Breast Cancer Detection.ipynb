#                          working on the dataset

# splitting the dataset into dependent and independent variables

```python
X=df.iloc[:,2:].values
print(X)
Y=df.iloc[:,1].values
#print(Y)
```

# Splitting data into training and testing data

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.20,random_state=0)
print(X_train)
```

# Feature scaling

```python
from sklearn.preprocessing import StandardScaler
X_train=StandardScaler().fit_transform(X_train)
X_test=StandardScaler().fit_transform(X_test)
print(X_train)
```

```python
'''from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=0)
lr.fit(X_train,Y_train)
print(lr.score(X_train,Y_train))'''
```

# MODELLING WITH LINEAR REGRESSION

Accuracy on training data

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
Yp_train=lr.predict(X_train)
from sklearn.metrics import r2_score
r2_score(Y_train,Yp_train)
```

The model does a good job (as it gives 77.70 % accuracy) on predicting with the training data

Accuracy on testing data

```python
lr.fit(X_test,Y_test)
Yp_test=lr.predict(X_test)
r2_score(Y_test,Yp_test)
```

Again the model did a good job (as it gives 83.61% accuracy)

# APPLYING LOGISTIC REGRESSION

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
log_r=LogisticRegression(random_state=0)
log_r.fit(X_train,Y_train)
Yp_log_r =log_r.predict(X_train)
print(accuracy_score(Y_train,Yp_log_r))
```
